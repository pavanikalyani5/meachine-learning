```python
## import the required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
#read the tain test data and the store.csv contain information related to stores which share the Primary key value that is common beteween train.csv 
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
store = pd.read_csv('store.csv')
```

```python
#Check if there are closed store with sales or any open stores with nosales
not_open = train[(train['Open'] == 0) & (train['Sales'] != 0)]
print("No closed store with sales: " + str(not_open.size == 0))

no_sales = train[(train['Open'] == 1) & (train['Sales'] <= 0)]
print("No open store with no sales: " + str(no_sales.size == 0))
```

```python
#Remove from training data all records where the sales is 0 since these are closed stores and these will not help in predicting Y
train = train.loc[train['Sales'] > 0]
```

```python
print('New Training Data Shape : ',train.shape)
```

```python
#When a csv file is imported and a Data Frame is made, the Date time objects in the file are read as a string object rather 
# a Date Time object and Hence it’s very tough to perform operations like Time difference on a string rather a Date Time object.
# Pandas to_datetime() method helps to convert string Date time into Python Date time object.
dates = pd.to_datetime(train['Date']).sort_values()
dates = dates.unique()
start_date = dates[0]
end_date = dates[-1]
print("Start date: ", start_date)
print("End Date: ", end_date)
date_range = pd.date_range(start_date, end_date).values
```

```python
# Visulize sales daywise we can observe that day7 sales is less
plt.rcParams['figure.figsize'] = (15.0, 12.0)
f, ax = plt.subplots(7, sharex=True, sharey=True)
for i in range(1, 8):
    data = train[train['DayOfWeek'] == i]
    ax[i - 1].set_title("Day {0}".format(i))
    ax[i - 1].scatter(data['Customers'], data['Sales'], label=i)

plt.legend()
plt.xlabel('Customers')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
#Assigining numbers to each categorical values fo rthe fileds Store Type,Assortment ans stateholiday
store['StoreType'] = store['StoreType'].astype('category').cat.codes
store['Assortment'] = store['Assortment'].astype('category').cat.codes
train["StateHoliday"] = train["StateHoliday"].astype('category').cat.codes
store.head()
```

```python
train.head()
```

```python
#merge the train as store dataframes having storeid as primary and foreign key relationship merging from left
merged = pd.merge(train, store, on='Store', how='left')
merged.head()
```

```python
#observe that the number of features have increased
merged.shape
```

```python
merged.isnull().sum()
```

```python
# replace missing values to zero
merged.fillna(0, inplace=True)
```

```python
#convert the date which is in strings to date time feature
merged['Date'] = pd.to_datetime(merged['Date'])
merged.dtypes
```

```python
#extracting 4 features from the datetime seperately
merged['Year'] = merged.Date.dt.year
merged['Month'] = merged.Date.dt.month
merged['Day'] = merged.Date.dt.day
merged['Week'] = merged.Date.dt.week
merged.head()
```

```python
# Number of months that competition has existed near a particular store
merged['MonthsCompetitionOpen'] = 12 * (merged['Year'] - merged['CompetitionOpenSinceYear']) + (merged['Month'] - merged['CompetitionOpenSinceMonth'])
merged.loc[merged['CompetitionOpenSinceYear'] == 0, 'MonthsCompetitionOpen'] = 0
```

```python
# Number of weeks that promotion has existed near a particular store
merged['WeeksPromoOpen'] = 12 * (merged['Year'] - merged['Promo2SinceYear']) + (merged['Date'].dt.weekofyear - merged['Promo2SinceWeek'])
merged.loc[merged['Promo2SinceYear'] == 0, 'WeeksPromoOpen'] = 0
```

```python
merged.dtypes
```

```python
#changing floats to int
toInt = [
        'CompetitionOpenSinceMonth',
        'CompetitionOpenSinceYear',
        'Promo2SinceWeek', 
        'Promo2SinceYear', 
        'MonthsCompetitionOpen', 
        'WeeksPromoOpen']

merged[toInt] = merged[toInt].astype(int)
```

```python
#compute median sales instead of Mean
med_store = train.groupby('Store')[['Sales', 'Customers', 'SalesPerCustomer']].median()
med_store.rename(columns=lambda x: 'Med' + x, inplace=True)

store = pd.merge(med_store.reset_index(), store, on='Store')
```
